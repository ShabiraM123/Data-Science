Count the Employee in each department

```python
Count_Emp=data.groupby('Department')['Department'].count()
```

```python
Count_Emp
```

```python
explode=[0,0.1,0]
palette_color=sns.color_palette('bright')
label=['Human Resources','Research & Development','Sales']
plt.pie(Count_Emp,autopct='%.0f%%',explode=explode,pctdistance=0.85,labels=label)
plt.title('Employee Distribution by Department')
```

Average salary of employees in each department

```python
Avg_Salary=data.groupby('Department')['MonthlyIncome'].mean()
```

```python
Avg_Salary.round(2)
```

```python
explode=[0,0.1,0]
palette_color=sns.color_palette('bright')
label=['Human Resources','Research & Development','Sales']
plt.pie(Avg_Salary,autopct='%.0f%%',explode=explode,pctdistance=0.85,labels=label)
plt.title(' Salary Distribution by Department')
```

Calculate the average of salary of employees based on gender

```python
Gender_Salary=data.groupby(['Department','Gender'],as_index=False).MonthlyIncome.mean()
```

```python
Gender_Salary.round(2)
```

```python
sns.barplot(x='Department',y='MonthlyIncome',hue='Gender',data=Gender_Salary)
```

Education Field of employees in each department

```python
Edu_Field=data.groupby(['Department','EducationField']).size().reset_index(name='Counts')
```

```python
Edu_Field
```

```python
sns.barplot(x='EducationField',y='Counts',hue='Department',data=Edu_Field)
```

Job Roles of employees in each department

```python
Job_Role=data.groupby(['Department','JobRole']).size().reset_index(name='Counts')
```

```python
Job_Role
```

```python
sns.barplot(x='JobRole',y='Counts',hue='Department',data=Job_Role)

plt.show()
```

Fetch the Employee Details by departments

```python
Employee_Details=data.groupby(['EmployeeNumber','Age','Education','Department','JobRole','TotalWorkingYears']).size().reset_index(name='Counts')
```
