# Data Analysis

To count the party in each state

```python
count_candidate=df.groupby('st_name')['cand_name'].size().reset_index(name='Counts')
```

```python
count_candidate
```

```python
plt.figure(figsize=(10, 6))
palette = sns.color_palette("Set1", n_colors=len('st_name'))
sns.barplot(x='st_name',y='Counts',data=count_candidate,palette=palette )
plt.xticks(fontsize=9)
plt.xticks(rotation=45)
plt.xlabel('States')
plt.ylabel('Candidate Numbers')
plt.title('candidate Numbers in Each State')
```

```python
poll_party=df.loc[:,['st_name','partyname','year','electors','totvotpoll']]
```

```python
poll_party
```

```python
sns.pairplot(data=poll_party)
```

```python
# Group the data by partyname and calculate the total votes for each party
party_votes = poll_party.groupby('partyname')['totvotpoll'].sum().reset_index()
```

```python
party_votes
```

```python
# Sort the parties by total votes in descending order
party_votes = party_votes.sort_values(by='totvotpoll', ascending=False)
```

```python
party_votes
```

```python
##plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
sns.barplot(data=party_votes, x='partyname', y='totvotpoll', palette='viridis', width=0.6)  # Adjust the width as needed

# Label your plot and axes
plt.xlabel('Party Name')
plt.ylabel('Total Votes')
plt.title('Total Votes by Party')
plt.show()
```
